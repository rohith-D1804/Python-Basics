DICTIONARIES BASICS

Create a dictionary

```python
student={"Name":"Rohith","Age":21,"Branch":"ECE"}
print(student)
```

Access Values

```python
marks = {"maths": 85, "physics": 78, "chemistry": 90}
phy_marks=marks["physics"]
print(phy_marks)
```

Add a key

```python
person={"Name":"Alex","Age":21,}
person["city"]="Chennai"
print(person)
```

Update a value

```python
price={"pen":10,"book":50}
price["book"]=60
print(price)
```

Delete a key with no error

```python
data = {"a": 1, "b": 2, "c": 3}
data.pop("b",None) # popped to avoid errors
print(data)
```

Print all Keys

```python
scores = {"A": 90, "B": 85, "C": 78}

for key in scores:
    print(key)
```

Print all values

```python
scores = {"A": 90, "B": 85, "C": 78}

for value in scores.values():
    print(value)
```

Print all key valu pairs

```python
items = {"apple": 100, "banana": 40, "mango": 60}
for keys,values in items.items():
    print(f"{keys}:{values}")
```

    Print count of dict

```python
employees = {"e1": "John", "e2": "Sam", "e3": "Raj"}
count=0
for i in employees:
    count+=1
print(count)
```

Check key existence

```python
user = {"username": "admin", "password": "1234"}
if "email" in user:
    print("Exists")
else:
    print("Not exists")
```
