SMART STUDENT PERFORMANCE ANALYZER

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
username = input("Enter username: ")
password = input("Enter password: ")

if username == name and password == "12345":

    subjects = int(input("Enter number of subjects: "))

    total = 0
    fail = False

    for i in range(1, subjects + 1):
        marks = int(input(f"Enter marks for subject {i}: "))
        total = total + marks

        if marks < 35:
            fail = True

    average = total / subjects

    if fail:
        result = "FAIL"
        grade = "F"
    else:
        result = "PASS"
        if average >= 90:
            grade = "A"
        elif average >= 75:
            grade = "B"
        elif average >= 60:
            grade = "C"
        elif average >= 50:
            grade = "D"
        else:
            grade = "F"

    if grade == "A" and average >= 90:
        scholarship = "Eligible"
    else:
        scholarship = "Not Eligible"

    maths = int(input("Enter Maths marks: "))
    phy = int(input("Enter Physics marks: "))
    che = int(input("Enter Chemistry marks: "))

    cutoff = maths + ((phy + che) / 2)

    if cutoff >= 180:
        college = "Tier A"
    elif cutoff >= 150:
        college = "Tier B"
    else:
        college = "Tier C and below"

    print("\n---------- STUDENT DETAILS ----------")
    print("Name:", name)
    print("Age:", age)
    print("Total Marks:", total)
    print("Average:", average)
    print("Result:", result)
    print("Grade:", grade)
    print("Cutoff:", cutoff)
    print("College Category:", college)
    print("Scholarship:", scholarship)

else:
    print("Invalid Credentials ")







```
