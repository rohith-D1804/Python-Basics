Even and odd set

```python
num=[1,2,3,4,5]
even_set=set()
odd_set=set()
for i in num:
    if i%2==0:
        even_set.add(i)
    if i%2!=0:
        odd_set.add(i)
print(f"Even numbers: {even_set}")
print(f"Odd numbers: {odd_set}")
```

Attendance system

```python
registered = {"ram", "sam", "tom", "raj"}
present = {"sam", "raj"}
print(f"Registered students: {registered}")
print(f"Presented students: {present}")
print(f"Absentees: {registered.difference(present)}")
```

High scorers

```python
marks = [95, 82, 67, 90, 55, 82]
total = 0
for i in marks:
    total += i
average = total / len(marks)
set_marks = set()
for i in marks:
    if i > average:
        set_marks.add(i)
print(set_marks)
```

Common characters

```python
s1 = "python"
s2 = "typhoon"
set_s1=set(s1)
set_s2=set(s2)
print(f"The common characters between python and typhoon is: {set_s1.intersection(set_s2)}")
```

Remove failed marks


```python
marks = {78, 45, 33, 60, 88}
passed_marks=set()
for i in marks:
    if i>=40:
        passed_marks.add(i)
print(passed_marks)
```

Students Grade 


```python
marks = [92, 76, 58, 89, 45]
grade_a=set(); grade_b=set(); grade_c=set()

for i in marks:
    if i>=85:
        grade_a.add(i)
    elif i>=70:
        grade_b.add(i)
    else:
        grade_c.add(i)
print(f"Grade-A: {grade_a}")
print(f"Grade-B: {grade_b}")
print(f"Grade-C: {grade_c}")
```

Unique words

```python
sentence = "python is easy and python is powerful"
unique_words=sentence.split()
words=set(unique_words)
print(words)
print(len(words))
```

Discounted price

```python
prices = [100, 250, 300, 150, 100]
discount=set()
for i in prices:
    amt=int(i*0.90)
    discount.add(amt)
print(discount)
```

Remove duplicates + total

```python
nums = [10, 20, 10, 30, 40]
total=0
unique_numbers=set(nums)
for i in unique_numbers:
    total+=i
print(f"Unique numbers: {unique_numbers}")
print(f"Sum of unique numbers: {total}")
```
