Condition Check

```python
n=int(input("Enter the number"))
if n>0:
    print(f"{n} is positive")
elif n<0:
    print(f" {n} is negative")
else:
    print(f" {n} is zero")
```

Even or Odd

```python
n=int(input("Enter the number"))

if n%2==0:
    print(f" {n} is even")
else:
    print(f" {n} is odd")
```

Simple Calculator

```python
a=int(input("Enter the number"))
b=int(input("Enter the number"))

symbols=input("Enter the operation +,-,*,/")

if symbols=="+":
    sum=a+b
    ans=sum
elif symbols=="-":
    diff=a-b
    ans=diff
elif symbols=="*":
    mul=a*b
    ans=mul
elif symbols=="/":
    div=a/b
    ans=div
print(ans)
```

Sum using loop

```python
n=int(input("Enter the number"))
sum=0
for i in range(1,n+1):
    sum+=i
print(sum)
```

Count even numbers

```python
n=int(input("Enter the number"))
count=0
for i in range(1,n+1):
    if i%2==0:
        count+=1
print(count)
```

Multiplication Table

```python
table=int(input("Enter the number"))
for i in range(1,11):
    tab=i*table
    print(f"{table}X{i}={tab}")
```
