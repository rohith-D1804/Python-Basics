Number Nature Analyzer

```python
num=int(input("Enter the number"))


if num%2==0 and num>0:
    print(f"{num} is positive even")
elif num>0:
    print(f"{num} is positive odd")
elif num<0:
    print(f"{num} is negative")
else:
    print('Zero')
```

Shop Discount System

```python
product_price=int(input("Enter the price"))

if product_price>=5000:
    discount_price=product_price*0.30
    product_price=product_price-discount_price
elif product_price>=3000:
    discount_price=product_price*0.20
    product_price=product_price-discount_price
elif product_price>=1000:
    discount_price=product_price*0.10
    product_price=product_price-discount_price
else:
    print("no discount!!")

print(f"You purchased for {product_price}")
```

Exam result logic

```python

maths = int(input("Enter Maths marks: "))
phy = int(input("Enter Physics marks: "))
chem = int(input("Enter Chemistry marks: "))
average_marks = (maths + phy + chem) / 3
if maths < 40 or phy < 40 or chem < 40:
    print(f"The student has FAILED and secured {average_marks}%")
elif average_marks >= 75:
    print(f"The student PASSED with DISTINCTION and secured {average_marks}%")
else:
    print(f"The student PASSED and secured {average_marks}%")

```

City Utility & Access System.

```python
# LOGIN SYSTEM
username = input("Enter username: ")
password = input("Enter password: ")

if username == "laptop" and password == "123":

    # INPUTS
    name = input("Enter your name: ")
    age = int(input("Enter your age: "))
    gender = input("Enter your gender (M/F): ")
    income = int(input("Enter your monthly income: "))
    units = int(input("Enter electricity units consumed: "))

    # AGE CATEGORY
    if age < 18:
        age_category = "Minor"
    elif age < 60:
        age_category = "Adult"
    else:
        age_category = "Senior Citizen"

    # ELECTRICITY BILL
    if units <= 100:
        bill = units * 1
    elif units <= 200:
        bill = (100 * 1) + (units - 100) * 2
    else:
        bill = (100 * 1) + (100 * 2) + (units - 200) * 3

    # DISCOUNT
    if age_category == "Senior Citizen":
        discount = bill * 0.20
    elif income < 25000:
        discount = bill * 0.10
    else:
        discount = 0

    final_bill = bill - discount

    # SERVICE ELIGIBILITY
    if age_category == "Adult" and income >= 30000:
        service = "Premium Services Enabled"
    else:
        service = "Basic Services Enabled"

    
    print("CITY UTILITY SUMMARY ")
    print("Name:", name)
    print("Age Category:", age_category)
    print("Gender:", gender)
    print("Income:", income)
    print("Units Consumed:", units)
    print("Original Bill:", bill)
    print("Discount Applied:", discount)
    print("Final Bill:", final_bill)
    print("Service Type:", service)

else:
    print("Access Denied")
```
