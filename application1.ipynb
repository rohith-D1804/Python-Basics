Budget Tracker

PSEUDO CODE

START

Input income

Create empty dictionary for expenses

Input number of categories

Repeat for each category:
    Input category name
    Input amount
    Store in dictionary

Set total_expense = 0

For each amount in dictionary:
    Add to total_expense

Calculate savings

Find:
    highest expense category
    lowest expense category

If savings > 0:
    budget status = "Good"
Else:
    budget status = "Over Budget"

Print report

END


```python
income=int(input("Enter your income"))
expenses={}
n=int(input("Enter the number of expenses categories"))
for i in range(n):
    category=input("Enter the category name")
    amount=int(input("Enter the amount spent in particular category"))
    expenses[category]=amount

total_expenses=0
for amt in expenses.values():
    total_expenses+=amt

savings=income-total_expenses

highest_amt=None
highest_cat=None
lowest_cat=None
lowest_amt=None
for cat,amt in expenses.items():
    if highest_cat is None or amt>highest_amt:
        highest_amt=amt
        highest_cat=cat
    if lowest_cat is None or amt<lowest_amt:
        lowest_amt=amt
        lowest_cat=cat

if savings > 0:
    budget_status = "Good"
else:
    budget_status = "Over Budget"


print("BUDGET REPORT")
print(f"Income: {income}")
print(f"Expenses: {expenses}")
print("Total Expenses:", total_expenses)
print("Savings:", savings)
print("Highest Expense:", highest_cat, "-", highest_amt)
print("Lowest Expense:", lowest_cat, "-", lowest_amt)
print(f"Budget Status:, {budget_status}")

    
```
