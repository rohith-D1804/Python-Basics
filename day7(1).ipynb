TRANSFORM PATTERN

Square of elements

```python
lst1=[1,2,3]
lst2=[]
for i in lst1:
    square=i**2
    lst2.append(square)
print(lst2)
```

Replace even numbers with 0

```python
lst=[2,1,3,4,6,5,9,7,8]

for i in lst:
    if i%2==0:
        lst[i]=0
print(lst)
    
```

Pass or Fail status

```python
marks=[78, 35, 20, 90]
result=[]
for i in marks:
    if i>=35:
        result.append("pass")
    else:
        result.append("fail")
print(result)
```

Temperature Conversion

```python
celsius=[0,20,30]
fahrenheit=[]
for c in celsius:
    f=(c*9/5)+32
    fahrenheit.append(f)
print(fahrenheit)
```

Salary adjustment system

```python
salary = [18000, 25000, 12000, 40000]
adjusted_salary=[]
for amt in salary:
    if amt<20000:
        new_salary=amt*1.20
    else:
        new_salary=amt*1.10
    adjusted_salary.append(int(new_salary))
print(adjusted_salary)
```

Double every number

```python
lst=[1,2,3,4]
lst1=[]
for i in lst:
    product=i*2
    lst1.append(product)
print(lst1)
```

Convert all numbers to negative

```python
lst=[3,-5,6,-4]
lst1=[]
for num in lst:
    if num>0:
        lst1.append(-num)
    else:
        lst1.append(num)
       
print(lst1)
```

Add 5 to each number

```python
num=[10,20,30]
num5=[]
for i in num:
    sum=i+5
    num5.append(sum)
print(num5)
```

Replace odd numbers with -1

```python
num=[1,2,3,4,5]
num_odd=[]
for i in num:
    if i%2!=0:
        num_odd.append(-1)
    else:
        num_odd.append(i)
print(num_odd)
```

Marks-Grades

```python
marks=[85, 72, 40, 30]
grades=[]
for i in marks:
    if i>=80:
        grades.append('A')
    elif i>=60:
        grades.append('B')
    elif i>=40:
        grades.append('P')
    else:
        grades.append('F')
print(grades)
```

Temperature Status

```python
temperature=[15,28,4]
status=[]
for i in temperature:
    if i<20:
        status.append("cold")
    elif i<=35:
        status.append("Warm")
    else:
        status.append("Hot")
print(status)
```

Mark odd or even for a number


```python
num=[1, 2, 3, 4]
char=[]
for i in num:
    if i%2==0:
        char.append("Even")
    else:
        char.append("Odd")
print(char)
```

Password Strength

```python
num=[2,8,19]
strength=[]
for i in num:
    if i<6:
        strength.append("Weak")
    elif i<10:
        strength.append("Medium")
    else:
        strength.append("Strong")
print(strength)
```

Discounted price

```python
price=[100,300,600]
discount_price=[]
for money in price:
    if money<200:
        disc=money*0.05
        final_price=money-disc
        discount_price.append(final_price)
    else:
        disc=money*0.1
        final_price=money-disc
        discount_price.append(final_price)
print(discount_price)
```
