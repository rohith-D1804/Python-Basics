## COMBINATION 1

Mark and result

```python
marks=(90,89,43,29,68)
result=[]
for mark in marks:
    if mark>=40:
        result.append("PASS")
    else:
        result.append("FAIL")
print(result)
```

Removing Duplicates in result

```python
actual_result=set(result)
print(actual_result)
```

Creating dictionary

```python
mark_result={}
mark_result=dict(zip(marks,result))
print(mark_result)
```

Expense analysis


```python
expenses=(300,400,500,900)
total=0
for i in expenses:
    total+=i
    highest_exp=max(expenses)
    lowest_exp=min(expenses)
print("Total expenses:",total)
print("Highest Expense:",highest_exp)
print("Lowest Expense:", lowest_exp)
```

Word Length Tracker

```python
word=("WORD")
count=0
num=[]
tot={}
for i in word:
    count+=1
    num.append(count)
    tot=dict(zip(word,num))
print(tot)
```

Even & Odd split

```python
lst_even=[]
lst_odd=[]
for i in range(1,21):
    if i%2==0:
        lst_even.append(i)
    else:
        lst_odd.append(i)
print( "List of even numbers:",lst_even)
print("List of odd numbers:" ,lst_odd)
```

Student Grade System

```python
marks=(90,70,60,50,40,30)
total=0
grade=[]
count=0
for i in marks:
    total+=i
    count+=1
    

    if i>=90:
        grade.append("A")
    elif i>=50:
        grade.append("B")
    else:
        grade.append("C")
avg=total/count
print(f"Total marks: {total}, Average:{avg}, Grade= {grade}")
```
