Student Performance Analyzer

```python
marks = [78, 45, 92, 33, 60, 88, 41]

total = 0
count = 0
count_pass = 0
count_fail = 0
pass_sum = 0

highest = marks[0]
lowest = marks[0]

for m in marks:
    total += m
    count += 1

    if m >= 40:
        count_pass += 1
        pass_sum += m
    else:
        count_fail += 1

    if m > highest:
        highest = m

    if m < lowest:
        lowest = m

average = total / count
pass_percentage = (count_pass / count) * 100

print("Total Marks:", total)
print("Average Marks:", round(average, 2))
print("Passed Students:", count_pass)
print("Failed Students:", count_fail)
print("Highest Mark:", highest)
print("Lowest Mark:", lowest)
print("Sum of Passed Marks:", pass_sum)
print("Pass Percentage:", round(pass_percentage, 2), "%")
```
