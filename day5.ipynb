WHILE LOOP


Print numbers from 1-10

```python
a=1
while (a<=10):
    print(a)
    a=a+1
```

print numbers reverse

```python
num=10
while (num>=1):
    print(num)
    num=num-1

```

Print Even numbers

```python
num=1
while (num<=20):
        if num%2==0:
            
            print(num)
        num=num+1
    
```

Sum of numbers

```python
n=int(input("Enter the number"))
sum=0
a=1
while (a<=n):
    sum=sum+a
    a=a+1
print(sum)

       
```

count odd numbers

```python
n = int(input("Enter the number: "))
count = 0
i = 1
while i <= n:
    if i % 2 != 0:
        count += 1
    i += 1
print("Number of odd numbers:", count)
```

Multiplication table

```python
n=int(input("Enter the number"))
i=1
while (i<=10):
    value=i
    i=i+1
    mul=value*n
    print(f"{n}X{value}={mul}")
```

Factorial

```python
n=int(input("Enter the number"))
fact=1
i=1
while(i<=n):
    fact=fact*i
    i=i+1
print(fact)
```

Largest even number

```python
n = int(input("Enter the number: "))

i = n
while i >= 1:
    if i % 2 == 0:
        print("Largest even number is:", i)
        break
    i -= 1
```
