Print pass or fail

```python
marks={"maths":78,"physics":35,"chemistry":84}
count=0
for key,values in marks.items():
    if values<40:
        print(f"{key}: FAIL")
    else:
        count+=1
        print(f"{key}: PASS")
print(f"{count} students were passed")
```

```python
marks = {"A": 65, "B": 82, "C": 74}
highest_mark = 0
highest_subject = ""
for subject, mark in marks.items():
    if mark > highest_mark:
        highest_mark = mark
        highest_subject = subject
print(f"Highest marks: {highest_mark} in subject {highest_subject}")
```

Add bonus 5 marks for each subject

```python
marks = {"maths": 80, "physics": 70}
for subject in marks:
    marks[subject]+=5
print(marks)
```

Seperate pass & fail

```python
marks = {"maths": 78, "physics": 35, "chemistry": 90}
passed_subject=[]
failed_subject=[]
for subject,mark in marks.items():
    if mark>=40:
        passed_subject.append(mark)
    else:
        failed_subject.append(mark)
print(f"Passed Subjects: {passed_subject}")
print(f"Failed Subjects: {failed_subject}")
```

Total and average marks

```python
marks = {"maths": 78, "physics": 35, "chemistry": 90}
num_subjects=len(marks)
passed_subject=[]
failed_subject=[]
total=0
for subject,mark in marks.items():
    total+=mark
    if mark>=40:
        passed_subject.append(mark)
    else:
        failed_subject.append(mark)
average_marks=int(total/num_subjects)
print(f"Total marks:{total}, Average: {average_marks}")
print(f"Passed Subjects: {passed_subject}")
print(f"Failed Subjects: {failed_subject}")
```

Checking value existing

```python
prices = {"pen": 10, "book": 50, "eraser": 5}
expensive_found=False
for price in prices.values():

    if price>40:
        expensive_found=True
        break
        
if expensive_found:
    print("Expensive items exists")
else:
    print("No expensive items")
```

Conert list to dict


```python
subjects = ["maths", "physics", "chemistry"]
marks = [80, 75, 90]

result = {}

for i in range(len(subjects)):
    result[subjects[i]] = marks[i]

print(result)
```

Update failed subject to 40

```python
marks = {"maths": 78, "physics": 35, "chemistry": 90}
for subject,mark in marks.items():
    if mark<=40:
        marks[subject]=40
print(marks)
```
