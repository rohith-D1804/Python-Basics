SET BASICS

Create a set

```python
set={10,20,30,40}
print(set)
print(len(set))
print(type(set))
```

Add an element

```python
set_1={1,2,3,4}
set_1.add(5)
print(set_1)
```

Remove an element

```python
colors={"green","blue","red"}
colors.remove("green")
print(colors)
```

Check membership

```python
a={1,2,3,15,20}
if 13 in a:
    print("15 is present")
else:
    print("15 is not present")
```

Count elements

```python
data={1,2,3,3,4}
count=0
for i in data:
    count+=1
print(f"There are {count} elements in {data}")
```

Remove duplicates

```python
lst = [1, 2, 2, 3, 3, 4]
set(lst)
```

Common Elements

```python
a={1,2,3,4,5}
b={3,4,5,6,7}
a.intersection(b)
#unique elements

print(f"Common elements: {a.intersection(b)}")
print(f"unique_elements: {a.union(b)}")
```

Difference

```python
a={10,20,20,30}
b={20,30,40,50}
print(f"b difference a: {b.difference(a)}") # elements in "b" will not be in "a"
print(f"a difference b: {a.difference(b)}") # elements in "a"  will not be in "b"
```

Even number in set

```python
a=set()
for  i in range(1,11):
    if i%2==0:
        a.add(i)
print(a)
```

Remove without causing error

```python
fruits={"banana","orange", "apple"}
fruits.discard("watermelon")
print(fruits)
```

Even and Odd numbers in seperate sets

```python
even = set()
odd = set()

for i in range(1, 21):
    if i%2==0:
        even.add(i)
    else:
        odd.add(i)

print("Even:", even)
print("Odd:", odd)
```
