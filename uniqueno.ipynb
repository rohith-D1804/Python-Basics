Unique Even numbers

```python
lst=[1,2,2,3,4,5,4,6,7,6,6,10]
even_set=set()
for i in lst:
    if i%2==0:
        even_set.add(i)
print(even_set)
```

Count common elements

```python
a={1,2,3}
b={2,3,4}
count=(len(a.intersection(b)))
print(count)
```

Remove duplicates and sort


```python
nums = [9, 1, 5, 3, 9, 1, 7]
s=set(nums)
sorted_set=sorted(s)
print(sorted_set)
```

Find which numbers are missing from present

```python
full={1,2,3,4,5,6}
present={2,3,5}
missing_number=full.difference(present)
print(missing_number)
```

Students list

```python
passed = {"A", "B", "C"}
failed = {"C", "D", "E"}

print(f"Passed students: {passed}")
print(f"failed students: {failed}")
print(f"Both the students: {passed.intersection(failed)}")
```

Remove all odd numbers


```python
nums = {1, 2, 3, 4, 5, 6, 7, 8}

nums = {n for n in nums if n % 2 == 0}

print(nums)

```

Common vowels

```python
s1 = "education"
s2 = "automation"
vowels = {'a', 'e', 'i', 'o', 'u'}
set1 = set(s1)
set2 = set(s2)
common_vowels = vowels.intersection(set1, set2)
print("Common vowels:", common_vowels)

```

Numbers divisible by 3 0r 5

```python
s=set()
for i in range(1,31):
    if i%3==0 and i%5==0:
        s.add(i)
print(f"The number between 1-30 which are divisible by 3 and 5 are:{s}")
print(f"The length: {len(s)}")
```

Students who did not attend

```python
registered = {"john", "alice", "bob", "david"}
attended = {"alice", "bob"}
print(f"Student did not attended: {registered.difference(attended)}")
print(f"Students attended: {registered.intersection(attended)}")
```
