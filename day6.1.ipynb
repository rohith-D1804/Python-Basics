Find smallest number in list

```python
lst=[1,2,3,4,0.5,6,0.1]
smallest=lst[0]
for i in lst:
    if i<smallest:
        smallest=i
print(smallest)
```

Create list of even numbers

```python
lst=[]
for i in range(1,11):
    lst.append(i)
lst_even=[]
for a in lst:
    if a%2==0:
        lst_even.append(a)
print(lst_even)
```

Count positive and negative number

```python
lst=[1,2,3,-3,-2,-1,-5,8,9,0,-9,-8]
count_positive=0
count_negative=0

for i in lst:
    if i<0:
        count_negative+=1
    elif i>0:
        count_positive+=1
print(f"There are {count_positive} positive numbers")
print(f"There are {count_negative} negative numbers")
```

Replace all even numbers with 0

```python
lst=[]
for i in range(1,11):
    lst.append(i)
for i in range (len(lst)):
    if lst[i]%2==0:
        lst[i]=0         
print(lst)
```
