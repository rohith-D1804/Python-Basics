## Basic application 


Student performance analyzer

```python
number_of_students=int(input("Enter the number of students"))
list_of_students={}


for i in range(number_of_students):
    name=input("Enter student name:")
    mark=int(input("Enter the marks"))
    list_of_students[name]=mark

def avg_marks():
    no_of_sub=len(list_of_students)
    total=0
    for name,mark in list_of_students.items():
        total+=mark
        total_mark=total
        if mark<=40:
            print(f"{name} is failed and secured {mark}")
    passed_students=dict(filter(lambda x:x[1]>40,list_of_students.items()))
    avg=total_mark/no_of_sub        
    print(f" Total marks of whole class: {total_mark}, Average of whole class: {avg}")  
     

    print(passed_students)
    print(list_of_students)
avg_marks()
```
