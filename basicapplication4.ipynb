## Login attempt analyzer

```python
def login_analyzer(attempts):
    report = {}

    report["total_attempts"] = len(attempts)
    report["failed_attempts"] = attempts.count("fail")

    if report["failed_attempts"] > 3:
        report["status"] = "Risky"
    else:
        report["status"] = "Safe"

    return report

attempts = ["success", "fail", "fail", "success", "fail", "fail"]
result = login_analyzer(attempts)

print("Total Attempts:", result["total_attempts"])
print("Failed Attempts:", result["failed_attempts"])
print("Status:", result["status"])

```
