## Duplicate Detector


Start

    Input how many numbers in list
    Input numbers in list
    Find duplicates in list
    Store duplicates in a seperate list
    def duplicate_count():
        count the duplicates
        return the value
End

```python
n=int(input("Enter the number of numbers to be added in list"))
num_lst=[]


for i in range(n):
    num=int(input("Enter the numbers"))
    num_lst.append(num)

seen=set()
dup=[]

for num in num_lst:
     if num in seen:
          dup.append(num)
     else:
          seen.add(num)


#function for counting duplicates
def duplicate_count():
     return len(dup)

print(f"Number in List: {num_lst}")
print(f"Duplicates in List: {dup}")
duplicate_count()

```
