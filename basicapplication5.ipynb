## Chat Moderation System

```python
n = int(input("Enter the number of messages: "))
message_lst = []

restricted_words = {"bad", "idiot", "stupid", "crackhead"}  
flagged_messages = set()  


for i in range(n):
    message = input("Enter your message: ")
    message_lst.append(message)

''
def moderate_messages():
    for msg in message_lst:
        words = msg.lower().split()   
        status = "Safe"

        for word in words:
            if word in restricted_words:
                status = "Flagged"
                flagged_messages.add(msg)
                break

        print(f"Message: {msg}")
        print(f"Status: {status}\n")


moderate_messages()

print("Flagged Messages:")
print(flagged_messages)


```
