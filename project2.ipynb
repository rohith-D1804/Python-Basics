STUDENT RESULT & FEE ANALYZER SYSTEM

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

username = input("Enter your username: ")
password = int(input("Enter your password: "))

if username == name and password == 123:

    no_of_subjects = int(input("Enter total number of subjects: "))

    marks = []
    for i in range(no_of_subjects):
        m = int(input("Enter marks: "))
        marks.append(m)

    # ACCUMULATORS
    total = 0
    count_pass = 0
    count_fail = 0
    highest = marks[0]
    lowest = marks[0]
    unique_marks=set(marks)
    passed_marks=set()
    cate={}

    result = []

    for n in marks:
        total += n

        if n >= 40:
            count_pass += 1
            passed_marks.add(n)
            result.append("PASS")
        else:
            count_fail += 1
            result.append("FAIL")

        if n > highest:
            highest = n
        if n < lowest:
            lowest = n

    average = total / no_of_subjects

    # GRADE & FEE
    base_price = 50000

    if average >= 85:
        grade = "A"
        fee_amount = base_price * 0.80
    elif average >= 70:
        grade = "B"
        fee_amount = base_price * 0.90
    elif average >= 50:
        grade = "C"
        fee_amount = base_price
    else:
        grade = "F"
        fee_amount = base_price

    cate[tuple(marks)]=result
    print("\n-------- STUDENT REPORT --------")
    print("Name:", name)
    print("Age:", age)
    print("Subjects:", no_of_subjects)
    print("Marks:", marks)
    print("Result:", result)
    print("Total Marks:", total)
    print("Average:", average)
    print("Passed:", count_pass)
    print("Failed:", count_fail)
    print("Highest:", highest)
    print("Lowest:", lowest)
    print("Grade:", grade)
    print("Final Fee:", fee_amount)
    print("Unique marks:", unique_marks)
    print("Passed marks:", passed_marks)
    print("Marks,Result:", cate)
else:
    print("Access Denied")
```
