Count numbers with condition

```python
n=int(input("Enter the number"))
count_2=0
count_3=0
for i in range(1,n+1):
    if i%2==0:
        count_2=count_2+1
    if i%3==0:
        count_3=count_3+1
print(f"There are {count_2} numbers are between 1-{n} which are divisible by 2")
print(f"There are {count_3} numbers are between 1-{n} which are divisible by 3")
```

Sum of squares

```python
n=int(input("Enter the number"))
total=0
for i in range(1,n+1):
    power=i**2
    total=total+power
print(f"Sum of squares between 1-{n} is: {total}")
```

First divisible number

```python
n=int(input("Enter the number"))
for i in range(1,n+1):
    if i%7==0:
        print(f"First number divisible by 7 between 1-{n} is :{i}")
        break
```

Sum except multiples

```python
n=int(input("Enter the number"))
for i in range(1,n+1):
    if i%5==0:
        continue
    print(i)
```

Greatest divisor


```python
n = int(input("Enter the number: "))

for i in range(n - 1, 0, -1):
    if n % i == 0:
        print("Greatest number less than", n, "that divides it is:", i)
        break
```

Number classification

```python
n = int(input("Enter the number: "))

even_count = 0
odd_count = 0
div3_count = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

    if i % 3 == 0:
        div3_count += 1

print("Even numbers:", even_count)
print("Odd numbers:", odd_count)
print("Divisible by 3:", div3_count)

    
```
