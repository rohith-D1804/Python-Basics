Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.

```python
array = np.random.randint(1, 21, size=(5, 5))
print("Original array: \n",array)
border_elements = np.concatenate((array[0, :], array[-1, :], array[1:-1, 0], array[1:-1, -1]))
print("Border elements: \n",border_elements)
```

Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.

```python
array1 = np.random.randint(1,5 , size=(3, 4))
array2 = np.random.randint(1,5 , size=(3, 4))
print("Original array1: \n", array1)
print("Original array2: \n", array2)

add_arr=array1+array2
sub_arr=array1-array2
mul_arr=array1*array2
div_arr=array1/array2
print("Addition: \n",add_arr)
print("Subtraction: \n",sub_arr)
print("Multiplication: \n",mul_arr)
print("Division: \n",div_arr)
```

Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.

```python
arra = np.random.randint(1,16 , size=(4, 4))
row_sum=np.sum(arra, axis=1)
col_sum=np.sum(arra, axis=0)
print("Row wise sum: \n", row_sum)
print("column wise sum: \n", col_sum)
```

 Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
array = np.arange(1, 10).reshape((3, 3))
print("Original array: \n",array)
mean = np.mean(array)
std_dev = np.std(array)
normalized_array = (array - mean) / std_dev
print("Normalized array: \n",normalized_array)
```
