LIST ELEMENTS

Create and print a list

```python
lst=[1,2,3,4,5,6,7]
print(lst)
print(len(lst))
```

Appending in list

```python
fruits=["apple","banana","mango"]
fruit=input("Enter your favourite fruit")
print(fruits)
fruits.append(fruit)
print(fruits)
```

Insert the number

```python
lst=[1,3,4,5,6,7,8,10]
lst.insert(1,2)
lst.insert(8,9)
print(lst)
```

Pop the list

```python
vegetables=["carrot","potato","beans"]
vegetables.pop()
print(vegetables)
```

Remove the vegetables from basket

```python
vegies=["carrot",'Potato',"Beans","Cauliflour"]
vegies.remove("Potato")
print(vegies)
```

length of the list

Sum of all elements in list

```python
lst=[1,2,3,4,5]
total=0
for i in lst:
    total+=i
print(total)
```

count of even numbers

```python
lst=[1,2,3,4,5]
count=0
for i in lst:
    if i%2==0:
        count+=1
print(count)
     
```

Largest number



```python
lst = [3, 7, 2, 9, 5]

largest = lst[0]

for i in lst:
    if i > largest:
        largest = i

print("Largest number:", largest)
```

Create new list with odd numbers

```python
lst_1=[]
lst_2=[1,2,3,4,5,6,7,8]
for i in lst_2:
    if i%2!=0:
        lst_1.append(i)
print(lst_1)
        
```
