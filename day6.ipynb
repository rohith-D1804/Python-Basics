LIST ELEMENTS

Creat and print a list

```python
lst=[1,2,3,4,5,6,7]
print(lst)
print(len(lst))
```

Appending in list

```python
fruits=["apple","banana","mango"]
fruit=input("Enter your favourite fruit")
print(fruits)
fruits.append(fruit)
print(fruits)
```

Insert the number

```python
lst=[1,3,4,5,6,7,8,10]
lst.insert(1,2)
lst.insert(8,9)
print(lst)
```

Pop the list

```python
vegetables=["carrot","potato","beans"]
vegetables.pop()
print(vegetables)
```

Remove the vegetables from basket

```python
vegies=["carrot",'Potato',"Beans","Cauliflour"]
vegies.remove["Potato"]
print(vegies)
```
