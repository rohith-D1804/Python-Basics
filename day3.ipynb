Count even numbers

```python
n=int(input("Enter the number"))
count=0
for i in range(0,n+1):
    if i%2==0:
        count=count+1
print(f"There are {count} even numbers between 0-{n}")
```

Sum of even & odd

```python
n=int(input("Enter the number"))
e_sum=0
o_sum=0
for i in range(1,n+1):
    if i%2==0:
        e_sum=e_sum+i
    else:
        o_sum=o_sum+i
print(f"The sum of even numbers={e_sum} and The sum of odd numbers={o_sum} between 1-{n}")
```

Prime number check

```python
n=int(input("Enter the number"))
if n<=1:
    print("Not a prime")
else:
    is_prime = True
for i in range(2,n):
    if n%i==0:
        is_prime= False
        break
if is_prime:
    print(f"The given number {n} is a prime number")
else:
    print(f"The given number {n} is not a prime number") 
```

Count numbers in range

```python
m=int(input("Enter the number"))
n=int(input("Enter the number"))
count_3=0
count_5=0
for i in range(m,n+1):
    if i%3==0:
        count_3=count_3+1
    if i%5==0:
        count_5=(count_5+1)
print(f"There are {count_3} numbers between {m} and {n} which are divisible by 3")
print(f"There are {count_5} numbers between {m} and {n} which are divisible by 5")
```
