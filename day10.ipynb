Access Values


```python
student = {"name": "Ravi", "marks": 85}
m=student["marks"]
print(m)
```

Add new entry

```python
student={"name":"Ravi","marks":85,}
student["Grade"]="A"
print(student)
```

Update value

```python
student={'name': 'Ravi', 'marks': 85, 'Grade': 'A'}
student["marks"]+=5
print(student)
```

Deleting without error


```python
student={'name': 'Ravi', 'marks': 85, 'Grade': 'A'}
student.pop("Grade")
print(student)
```

Print only keys

```python
marks = {"maths": 80, "physics": 70, "chemistry": 90}
for subject in marks.keys():
    print(subject)
    
```

Print all values

```python
marks = {"maths": 80, "physics": 70, "chemistry": 90}
for mark in marks.values():
    print(mark)
```

Print key:value

```python
marks = {"maths": 80, "physics": 70, "chemistry": 90}
for subject, mark in marks.items():
    print(f"{subject}:{mark}")
```

Pass or Fail

```python
marks = {"maths": 20, "physics": 76, "chemistry": 90}
for subject,mark in marks.items():
    if mark>=40:
        result="PASS"
    else:
        result="FAIL"
    print(f"{subject}:{result}")
```

Count passed subjects

```python
marks = {"maths": 60, "physics": 90,"chemistry": 35}
count=0
for mark in marks.values():
    if mark>=40:
        count+=1
print(f"The student is passed in {count} subjects")
```

Highest marks

```python
marks = {"maths": 20, "physics": 96, "chemistry": 90}
highest=0
for subject,mark in marks.items():
    if mark>highest:
        highest=mark

print(f"The highest score is :{highest}")
```

Bonus 5 to each subject

```python
marks = {"maths": 20, "physics": 76, "chemistry": 90}
for mark in marks:
    marks[mark]+=5
print(marks)
    
```

Price filter

```python
prices = {"pen": 10, "book": 50, "bag": 700}
for product,amt in prices.items():
    if amt>100:
        print(product)
        
```

Print names of absentees


```python
attendance = {"ram": "present", "sam": "absent", "raj": "present"}
for name,attend in attendance.items():
    if attend=="absent":
        print(name)
```

Build dictionary from input

```python
marks = {}

n = int(input("Enter number of subjects: "))

for i in range(n):
    subject = input("Enter subject name: ")
    mark = int(input("Enter mark: "))
    marks[subject] = mark

print("Final dictionary:", marks)
```
